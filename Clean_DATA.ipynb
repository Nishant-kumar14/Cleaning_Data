## 2. Handle missing values 

```python
data['Income'].fillna(data['Income'].median(), inplace=True)
```

```python
data.isnull().sum()/len(data)*100
```

## 3. Remove duplicates

```python
data.drop_duplicates(inplace=True)
```

## 4. Standardize text values

```python
data['Education'] = data['Education'].str.strip().str.title()
```

## 5. Convert date to consistent format

```python
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], dayfirst=True)
```

## 6. Rename columns to clean format

```python
data.columns = data.columns.str.lower().str.replace(" ", "_")
```

## 7. Check and fix data types

```python
data['year_birth'] = data['year_birth'].astype(int)
data['income'] = data['income'].astype(float)
```

## 8. Drop Columns

```python
data.drop(['id', 'z_costcontact', 'z_revenue'], axis=1, inplace=True)
```
